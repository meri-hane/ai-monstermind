# EASY MODE LOOK UP

```python
import itertools

# Define the colors and pegs
colors = ['B', 'G', 'P', 'R', 'Y', 'O']
num_pegs = 5

score_dict = {}
```

```python
all_guesses = list(itertools.product(colors, repeat=num_pegs))

# Print all guesses
all_guesses
```

```python
# Function to calculate hints
def calculate_hints(guess, answer):
    hints = []
    guess_no_match = []
    ans_no_match = []
    
    for guess_elem, ans_elem in zip(guess, answer):
        if guess_elem == ans_elem:
            hints.append("B")
        else:
            guess_no_match.append(guess_elem)
            ans_no_match.append(ans_elem)
    
    for guess_elem in guess_no_match:
        if guess_elem in ans_no_match:
            hints.append("W")
            ans_no_match.remove(guess_elem)
    
    return ''.join(hints)
```

```python
lookup_table = {}

initial_guess = ('B', 'B', 'B', 'G', 'G')  # Define the initial guess

for guess in all_guesses:
    guess_key = tuple(guess)
    lookup_table[guess_key] = []
    
    possible_answers = all_guesses.copy()
    
    # Remove the initial guess from possible answers
    if initial_guess in possible_answers:
        possible_answers.remove(initial_guess)
    
    current_answer = initial_guess  # Set the initial guess
    lookup_table[guess_key].append(''.join(current_answer))
    
    while possible_answers:
        if current_answer == guess:
            break
        
        new_possible_answers = []
        correct_hints = calculate_hints(current_answer, guess)
        
        for ans in possible_answers:
            if calculate_hints(current_answer, ans) == correct_hints:
                new_possible_answers.append(ans)
        
        possible_answers = new_possible_answers
        
        if possible_answers:  # Ensure there are possible answers left before popping
            current_answer = possible_answers.pop(0)
            lookup_table[guess_key].append(''.join(current_answer))
```

# EASY MODE ALGO Without Look Up

```python
from itertools import product

# Define the colors and pegs
colors = ['B', 'G', 'P', 'R', 'Y', 'O']
num_pegs = 5

# Function to calculate hints
def calculate_hints(guess, answer):
    hints = []
    guess_no_match = []
    ans_no_match = []
    
    for guess_elem, ans_elem in zip(guess, answer):
        if guess_elem == ans_elem:
            hints.append("B")
        else:
            guess_no_match.append(guess_elem)
            ans_no_match.append(ans_elem)
    
    for guess_elem in guess_no_match:
        if guess_elem in ans_no_match:
            hints.append("W")
            ans_no_match.remove(guess_elem)
    
    return ''.join(hints)
```

```python
# Define the secret answer
secret_answer = tuple('YOYG')  # Set the secret answer here

# Initialize the set of possible solutions
possible_solutions = set(product(colors, repeat=num_pegs))

# Make the initial guess (for simplicity, let's start with 'BBGG')
initial_guess = tuple('BBGG')
```

```python

# Iterate until the solution is found
while True:
    # Provide the guess to the user or the game
    print("Guess:", initial_guess)
    
    # Calculate feedback based on the secret answer
    feedback = calculate_hints(initial_guess, secret_answer)
    print("Feedback:", feedback)
    
    # Check if the guess is correct
    if feedback == 'BBBB':
        print("Found the solution:", initial_guess)
        break
    
    # Update the set of possible solutions based on the feedback
    possible_solutions = {
        solution for solution in possible_solutions
        if calculate_hints(initial_guess, solution) == feedback
    }
    
    # Choose the next guess based on the updated set of possible solutions
    initial_guess = possible_solutions.pop()
```

```python
# Function to guess the code
def guess_code(secret_answer):
    initial_guesses = [
        ('B', 'B', 'G', 'G', 'G'),
        ('B', 'B', 'G', 'G', 'R'),
        ('B', 'B', 'B', 'G', 'G')
    ]
    
    all_guesses_results = []
    
    for initial_guess in initial_guesses:
        possible_solutions = set(product(colors, repeat=num_pegs))
        guesses = ["".join(initial_guess)]
        
        while True:
            # Calculate feedback based on the secret answer
            feedback = calculate_hints(initial_guess, secret_answer)
            
            # Check if the guess is correct
            if feedback == 'BBBBB':
                break
            
            # Update the set of possible solutions based on the feedback
            possible_solutions = {
                solution for solution in possible_solutions
                if calculate_hints(initial_guess, solution) == feedback
            }
            
            # Choose the next guess based on the updated set of possible solutions
            if possible_solutions:
                initial_guess = possible_solutions.pop()
                guesses.append("".join(initial_guess))
            else:
                break
        
        all_guesses_results.append(guesses)
    
    # Find the list of guesses with the least number of attempts
    shortest_guesses = min(all_guesses_results, key=len)
    return shortest_guesses
```

```python
def guess_code(secret_answer, mode):
        if mode == 'average':
            initial_guesses = [
                ('B', 'B', 'G', 'G', 'G'),
                ('B', 'B', 'G', 'G', 'R'),
                ('B', 'B', 'B', 'G', 'G')
            ]
            pegs = 5
        else: 
            initial_guesses = [('B', 'B', 'G', 'G')]
            pegs = 4
        
        all_guesses_results = []
        
        for initial_guess in initial_guesses:
            possible_solutions = set(product(colors, repeat=pegs))
            guesses = ["".join(initial_guess)]
            
            while True:
                # Calculate feedback based on the secret answer
                feedback = calculate_hints(initial_guess, secret_answer)
                
                # Check if the guess is correct
                if feedback == 'B' * pegs:
                    break
                
                # Update the set of possible solutions based on the feedback
                possible_solutions = {
                    solution for solution in possible_solutions
                    if calculate_hints(initial_guess, solution) == feedback
                }
                
                # Choose the next guess based on the updated set of possible solutions
                if possible_solutions:
                    initial_guess = possible_solutions.pop()
                    guesses.append("".join(initial_guess))
                else:
                    break
            
            all_guesses_results.append(guesses)
        
        # Find the list of guesses with the least number of attempts
        shortest_guesses = min(all_guesses_results, key=len)
        return shortest_guesses
```
